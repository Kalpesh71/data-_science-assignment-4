2. Differentiate between:
● Series and DataFrame

● loc[] and iloc[] functions in Pandas

(a) Series and DataFrame

Series:
• One-dimensional data structure.

• Contains a single column of data.

• Created using pd.Series().

DataFrame:
• Two-dimensional data structure.

• Contains rows and columns.

• Created using pd.DataFrame().

Example:

import pandas as pd

s = pd.Series([10,20,30])
print(s)

df = pd.DataFrame({
"Name":["Amit","Neha"],
"Marks":[85,90]
})
print(df)

(b) loc[] and iloc[] Functions in Pandas

loc[]:
• Label-based indexing.

• Uses row and column names.

iloc[]:
• Integer position-based indexing.

• Uses row and column positions.

Example:

import pandas as pd

df = pd.DataFrame({
"Name":["Amit","Neha","Rahul"],
"Marks":[85,90,78]
})

print(df.loc[1,"Name"])
print(df.iloc[1,0])

3. Create a Pandas DataFrame using the dataset below:
Name Age Marks
Amit 20 85
Neha 21 90
Rahul 19 78
Sneha 22 88

a) Tasks:

● Display first 2 rows

● Display only Name and Marks columns

● Add a new column named “Result”

● Display students with marks greater than 80

b) Expected Concepts:
● pd.DataFrame()

● head()

● Column Selection

● Conditional Filtering

```python
import pandas as pd

data = {
    'Name': ['Amit', 'Neha', 'Rahul', 'Sneha'],
    'Age': [20, 21, 19, 22],
    'Marks': [85, 90, 78, 88]
}

df = pd.DataFrame(data)

print("Original DataFrame:")
print(df)
```

a)
 Tasks:
1. Display first 2 rows

```python
print(df.head(2))
```

2. Display only Name and Marks columns

```python
print(df[['Name', 'Marks']])
```

3. Add a new column named “Result”

```python
df['Result'] = df['Marks'].apply(lambda x: 'Pass' if x > 80 else 'Fail')
print(df)
```

4. Display students with marks greater than 80

```python
print(df[df['Marks'] > 80])
```

4. Create a DataFrame using the dataset below:

Name Age Salary

A 25 30000

B NaN 40000

C 27 NaN

D 24 35000

a)
 Tasks:

● Identify missing values

● Fill missing Age with average age

● Fill missing Salary with median salary

● Display cleaned dataset

b) Expected Concepts:

● isnull()

● fillna()

● mean()

● median()


```python
import pandas as pd
import numpy as np

data = {
    'Name': ['A', 'B', 'C', 'D'],
    'Age': [25, np.nan, 27, 24],
    'Salary': [30000, 40000, np.nan, 35000]
}

df = pd.DataFrame(data)

print("Original DataFrame:")
print(df)
```

```python
print("Missing Values in each column:")
print(df.isnull().sum())

print("\nMissing Values (Boolean Matrix):")
print(df.isnull())
```

```python
mean_age = df['Age'].mean()

df['Age'] = df['Age'].fillna(mean_age)

print("After filling missing Age with mean:")
print(df)
```

```python
median_salary = df['Salary'].median()

df['Salary'] = df['Salary'].fillna(median_salary)

print("\nAfter filling missing Salary with median:")
print(df)
```

```python
median_salary = df['Salary'].median()
df['Salary'] = df['Salary'].fillna(median_salary)

print("\nAfter filling missing Salary with median:")
print(df)
```

```python
import pandas as pd

# Load the Iris Dataset
url = "https://raw.githubusercontent.com/mwaskom/seaborn-data/master/iris.csv"
df = pd.read_csv(url)

print("Original Dataset (First 5 rows):")
print(df.head())
```

```python
print("Dataset Information:")
print(df.info())
```

```python
print(f"Total Rows: {df.shape[0]}")
print(f"Total Columns: {df.shape[1]}")
```

```python
avg_sepal = df['sepal_length'].mean()
print(f"Average Sepal Length: {avg_sepal:.2f} cm")
```

```python
print("Different Species:")
print(df['species'].unique())
```

```python
print("Summary Statistics:")
print(df.describe())
```
